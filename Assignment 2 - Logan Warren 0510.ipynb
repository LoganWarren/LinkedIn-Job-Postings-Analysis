## Look for Outliers:
- Looking into outliers in specific columns such as Total Applicants and LinkedIn Followers

#### Total_applicants:
- The Boxplots show that there are several vlaues that are far from the median that can suggest there are potential outliers
- The Histogram shows a right skew distribution, which shows there are a few extreamely high values in the column.

#### LinkedIn_Followers:
- The Boxplot shows that the followers on linkedin may have outliers
- LinkedIn's Histogram is also right skewed showing that there are some compaines with a much higher amount of followers than others.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram and Boxplot for Total_applicants:

sns.boxplot(x=df['Total_applicants'], ax=axes[0, 0])
axes[0, 0].set_title('Boxplot of Total Applicants')

sns.histplot(df['Total_applicants'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Histogram of Total Applicants')

# Histogram and Boxplot for LinkedIn_Followers: 

sns.boxplot(x=df['LinkedIn_Followers'], ax=axes[1, 0])
axes[1, 0].set_title('Boxplot of LinkedIn Followers')

sns.histplot(df['LinkedIn_Followers'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Histogram of LinkedIn Followers')
```

## Look at correlations:
From the Spearman Heatmap:
- There is a strong correlation between the Employee_count and LinkedIn_Followers
- There are two weak correlations, between Total_applicants and Employee_count and between Total_applicants and LinkedIn_followers

```python
spearman = df.corr(method="spearman", numeric_only=True)
sns.heatmap(spearman, annot=True)
```
